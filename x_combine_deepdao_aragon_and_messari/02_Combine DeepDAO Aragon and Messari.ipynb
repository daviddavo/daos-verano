```python
import pandas as pd

# load 01_messari_daos/messari_proposal_ids
messari = pd.read_csv('../01_messari_daos/messari_proposal_ids.csv')

# load 01_deepdao_daos/1687523727_deepdao_daos.csv
deepdao = pd.read_csv('../01_deepdao_daos/1687523727_deepdao_daos.csv')
```

```python
messari
```

```python
deepdao
```

```python
# deepdao drop platformId, name, deepdaoOrganizationId, deepdaoDaoId, mostRecentProposalDate, aum
deepdao = deepdao.drop(columns=['platformId', 'deepdaoOrganizationId', 'deepdaoDaoId', 'mostRecentProposalDate', 'aum'])
# rename platformTitle to platformType
deepdao = deepdao.rename(columns={'platformTitle': 'platformType'})
# remove duplicate websites but keep nulls
deepdao = deepdao.drop_duplicates(subset=['name', 'website'])
deepdao
```

```python
# messari set platformType to Snapshot if snapshot_id is not null
messari.loc[messari['snapshot_id'].notnull(), 'platformType'] = 'Snapshot'
# same for tally
messari.loc[messari['withtally_id'].notnull(), 'platformType'] = 'Tally'
# combine snapshot_id and withtally_id into platformId
messari['platformId'] = messari['snapshot_id'].fillna(messari['withtally_id'])
# drop snapshot_id and withtally_id
messari = messari.drop(columns=['snapshot_id', 'withtally_id'])
# drop count
messari = messari.drop(columns=['count'])
messari
```

```python
# extract snapshot id from website
import numpy as np

def get_snapshot_id(website):
    # format https://snapshot.org/#/<snapshot_id>/proposal/<proposal_id>
    if pd.isnull(website): return ''
    if 'snapshot.org/#/' in website:
        return website.split('snapshot.org/#/')[1].split('/')[0]
    else:
        return np.nan
    
deepdao['platformId'] = deepdao['website'].apply(get_snapshot_id)
deepdao
```

```python
# show deepdao rows with website null
deepdao[deepdao['website'].isnull()].groupby('platformType')['name'].count()
```

```python
# ignore columns with platformType 'Aragon' because we'll add all of the aragon daos later
deepdao = deepdao[deepdao['platformType'] != 'Aragon']
deepdao
```

## Pulling in Aragon data

```python
aragon_organizations = pd.read_csv('../daoanalyzer/organizations.csv')
aragon_organizations = aragon_organizations[['id', 'name']]
# add column platformType = Aragon
aragon_organizations['platformType'] = 'Aragon'
# rename id to platformId
aragon_organizations = aragon_organizations.rename(columns={'id': 'platformId'})
aragon_organizations
```

## Merging

```python
# outer join
combined = pd.merge(messari, deepdao, how='outer', on='platformId', suffixes=('_messari', '_deepdao'))
combined
```

```python
# desribe
combined.describe()
```

```python
# join the platform type columns, preserve both values if they are different
combined['platformType'] = combined['platformType_messari'].fillna(combined['platformType_deepdao'])
combined['platformTypeA'] = combined['platformType_messari'].fillna(combined['platformType_deepdao'])
combined['platformTypeB'] = combined['platformType_deepdao'].fillna(combined['platformType_messari'])
# where they are different, combine the values w comma
combined.loc[combined['platformTypeA'] != combined['platformTypeB'], 'platformType'] = combined['platformTypeA'] + ', ' + combined['platformTypeB']
# show where they are different
combined[combined['platformTypeA'] != combined['platformTypeB']]
```

```python
# remove platformTypeA and platformTypeB and platformType_messari and platformType_deepdao
combined = combined.drop(columns=['platformTypeA', 'platformTypeB', 'platformType_messari', 'platformType_deepdao'])
combined
```

```python
# remove duplicate rows
combined = combined.drop_duplicates(subset=['platformId'])
combined
```

```python
# if website is of snapshot.org, and it is missing the platformType, set the platformType to Snapshot
combined.loc[(combined['website'].str.contains('snapshot.org')) & (combined['platformType'].isnull()), 'platformType'] = 'Snapshot'

# if website is of snapshot.org and has no platformId, extract the snapshot id
combined.loc[(combined['website'].str.contains('snapshot.org')) & (combined['platformId'].isnull()), 'platformId'] = combined['website'].apply(get_snapshot_id)

combined
```

```python
# show rows where platformType or platformId is null
combined[combined['platformType'].isnull() | combined['platformId'].isnull()]
```

```python
# and remove em
combined = combined.dropna(subset=['platformType', 'platformId'])
```

```python
# drop website column
combined = combined.drop(columns=['website'])
```

```python
# drop dupe rows
combined = combined.drop_duplicates(subset=['platformId', 'platformType'])
combined.shape
```

```python
# reorder columns id, type, name
combined = combined[['platformId', 'platformType', 'name']]
combined
```

### Add in Aragon data

```python
# concat combined and aaraon_organizations
combined = pd.concat([combined, aragon_organizations])
combined
```

```python
combined.platformType.value_counts()
```

### Clean up strange values

```python
# find rows where platformType is not in ['Snapshot', 'Tally', 'Aragon']
combined[~combined['platformType'].isin(['Snapshot', 'Tally', 'Aragon'])]
```

```python
# make 'Snapshot, Safe/Snapshot' values be just snapshot
combined.loc[combined['platformType'] == 'Snapshot, Safe/Snapshot', 'platformType'] = 'Snapshot'
# make Safe/Snapshot be just snapshot
combined.loc[combined['platformType'] == 'Safe/Snapshot', 'platformType'] = 'Snapshot'
combined[~combined['platformType'].isin(['Snapshot', 'Tally', 'Aragon'])]
```

```python
# find the row with platformId rari and platformType 'Tally, Snapshot' and set it to 'Tally'
combined.loc[(combined['platformId'] == 'rari') & (combined['platformType'] == 'Tally, Snapshot'), 'platformType'] = 'Tally'
combined.loc[(combined['platformId'] == 'ampleforth') & (combined['platformType'] == 'Tally, Snapshot'), 'platformType'] = 'Tally'
combined[~combined['platformType'].isin(['Snapshot', 'Tally', 'Aragon'])]
```

```python
# remove duplicate rows
combined = combined.drop_duplicates(subset=['platformId', 'platformType'])
combined.shape
```
