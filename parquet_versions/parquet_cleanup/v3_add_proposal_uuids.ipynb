```python
IN_FILE = '../../parquets/proposals.parquet'
OUT_FILE = '../../parquets/proposals.parquet'
DEPLOYMENTS_FILE = '../../parquets/deployments.parquet'
```

```python
# read in ../parquets_version3/proposals.parquet
import pyarrow.parquet as pq

proposals = pq.read_table(IN_FILE)
proposals
```

```python
# make pandas
proposals = proposals.to_pandas()
proposals.head()
```

```python
# make col platform_deployment_id_to_hash <platform>_<platform_deployment_id>
proposals['platform_deployment_id_to_hash'] = proposals.apply(lambda row: row['platform'] + '_' + str(row['platform_deployment_id']), axis=1)
proposals.head()
```

```python
# uuid.uuid5(uuid.NAMESPACE_OID, str(ids_to_hash[i]))
import uuid
proposals['deployment_id'] = proposals.apply(lambda row: str(uuid.uuid5(uuid.NAMESPACE_OID, row['platform_deployment_id_to_hash'])), axis=1)
proposals.head()
```

```python
proposals.columns
```

```python
# make an id column that hashes <platform>_<platform_deployment_id>_<proposal_id>
proposals['id'] = proposals.apply(lambda row: str(uuid.uuid5(uuid.NAMESPACE_OID, row['platform'] + '_' + str(row['platform_deployment_id']) + '_' + str(row['proposal_id']))), axis=1)
proposals.head()
```

```python
proposals.columns
```

```python
proposals = proposals[[
    'id',
    # 'platform_deployment_id',
    # 'platform_deployment_id_to_hash',
    'deployment_id',
    'proposal_id', # rename to platform_proposal_id
    'author',
    'date',
    'votes_count',
    'platform',
    'title',
    'description',
    'start',
    'end',
]]
proposals.head()
```

```python
# confirm that the uuids are done properly by merging on the deployment_id with deploys

# read in ../parquets_version3/deploys.parquet as pandas
import pyarrow.parquet as pq
deploys_df = pq.read_table(DEPLOYMENTS_FILE).to_pandas()
deploys_df.head()
```

```python
# merge on deployment_id
proposals_merged = proposals.merge(deploys_df, left_on='deployment_id', right_on='id', how='left', validate='many_to_one')
# sort by deployment id
proposals_merged = proposals_merged.sort_values(by=['deployment_id'])
proposals_merged.head()
```

```python
assert not proposals_merged.empty
# looks good, merges properly
```

```python
# export to pyarrow
import pyarrow as pa
proposals = pa.Table.from_pandas(proposals)
proposals
```

```python
proposals.column_names
```

```python
# rename proposal_id to platform_proposal_id
proposals = proposals.rename_columns(['id',
  'deployment_id',
  'platform_proposal_id',
  'author',
  'date',
  'votes_count',
  'platform',
  'title',
  'description',
  'start',
  'end',
])
proposals
```

```python
proposals.schema
```

```python
# save
pq.write_table(proposals, OUT_FILE)
```
