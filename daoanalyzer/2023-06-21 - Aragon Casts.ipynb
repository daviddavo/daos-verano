```python
# load casts.csv into a dataframe
import pandas as pd
casts = pd.read_csv('casts.csv')
casts.shape
```

```python
# drop unnamed column
casts = casts.drop('Unnamed: 0', axis=1)
```

```python
# make createdAt epoch a datetime object
casts['createdAt'] = pd.to_datetime(casts['createdAt'], unit='s')
casts
```

```python
# get date a year ago
from datetime import datetime, timedelta
a_year_ago = datetime.now() - timedelta(days=365)

# filter casts to only include casts from the last year
this_years_casts = casts[casts['createdAt'] > a_year_ago]
this_years_casts
```

```python
# calculate unique voters per orgAddress
unique_voters_by_organization = this_years_casts.groupby('orgAddress')['voter'].nunique().sort_values(ascending=False)
unique_voters_by_organization
```

```python
token_holders = pd.read_csv('tokenHolders.csv')
token_holders.shape
```

```python
token_holders = token_holders.drop('Unnamed: 0', axis=1)
token_holders.head()
```

```python
# group by organizationAddress and count number of addresses
token_holders_by_organization = token_holders.groupby('organizationAddress').agg({'address': 'count'}).sort_values(by='address', ascending=False)
token_holders_by_organization
```

```python
# left join token_holders_by_organization and unique_voters_by_organization
# TODO: validate the merge with the validation param
organizations = pd.merge(token_holders_by_organization, unique_voters_by_organization, left_index=True, right_index=True, how='inner')
organizations
```

```python
# calculate the percentage of token holders who have voted on more than 10% of the proposals
# group this years casts by voter and organization and count the number of proposals they have voted on
voters = this_years_casts.groupby(['voter', 'orgAddress']).agg({'voteId': 'count'}).sort_values(by='voteId', ascending=False)
# rename voteId to vote_count
voters = voters.rename(columns={'voteId': 'vote_count'})
voters
```

```python
# get the number of proposals per organization
proposals_per_organization = this_years_casts.groupby('orgAddress').agg({'voteId': 'count'}).sort_values(by='voteId', ascending=False)
proposals_per_organization = proposals_per_organization.rename(columns={'voteId': 'vote_count'})
proposals_per_organization
```

```python
# join voters and proposals_per_organization
pct_voted_on_metrics = pd.merge(voters, proposals_per_organization, left_index=True, right_index=True, how='inner')
# rename vote_count_x to voted_on_count and vote_count_y to proposal_count
pct_voted_on_metrics = pct_voted_on_metrics.rename(columns={'vote_count_x': 'voted_on_count', 'vote_count_y': 'proposal_count'})
pct_voted_on_metrics
```

```python
# add column pct_voted_on
pct_voted_on_metrics['pct_voted_on'] = pct_voted_on_metrics['voted_on_count'] / pct_voted_on_metrics['proposal_count']
pct_voted_on_metrics
```

```python
# add a column that is true if pct_voted_on > 0.1
pct_voted_on_metrics['voted_on_more_than_10_pct_of_proposals'] = pct_voted_on_metrics['pct_voted_on'] > 0.1
# add a column that is true if pct_voted_on > 0.05
pct_voted_on_metrics['voted_on_more_than_5_pct_of_proposals'] = pct_voted_on_metrics['pct_voted_on'] > 0.05
# add a column that is true if pct_voted_on > 0.01
pct_voted_on_metrics['voted_on_more_than_1_pct_of_proposals'] = pct_voted_on_metrics['pct_voted_on'] > 0.01
pct_voted_on_metrics
```

```python
# group by orgAddress and keep the count of the voted_on_more_than_10_pct_of_proposals and voted_on_more_than_5_pct_of_proposals and voted_on_more_than_1_pct_of_proposals columns
pct_voted_on_metrics_by_organization = pct_voted_on_metrics.groupby('orgAddress').agg({'voted_on_more_than_10_pct_of_proposals': 'sum', 'voted_on_more_than_5_pct_of_proposals': 'sum', 'voted_on_more_than_1_pct_of_proposals': 'sum'})
pct_voted_on_metrics_by_organization
```

```python
organizations
```

```python
# join organizations and pct_voted_on_metrics_by_organization
organizations = pd.merge(organizations, pct_voted_on_metrics_by_organization, left_index=True, right_index=True, how='left', validate='one_to_one')
organizations
```

```python
# calculate percentage of token holders who have voted on more than 10% of the proposals
organizations['pct_of_token_holders_who_have_voted_on_more_than_10_pct_of_proposals'] = organizations['voted_on_more_than_10_pct_of_proposals'] / organizations['token_holders']
# calculate percentage of token holders who have voted on more than 5% of the proposals
organizations['pct_of_token_holders_who_have_voted_on_more_than_5_pct_of_proposals'] = organizations['voted_on_more_than_5_pct_of_proposals'] / organizations['token_holders']
# calculate percentage of token holders who have voted on more than 1% of the proposals
organizations['pct_of_token_holders_who_have_voted_on_more_than_1_pct_of_proposals'] = organizations['voted_on_more_than_1_pct_of_proposals'] / organizations['token_holders']
# drop the voted_on_more_than_10_pct_of_proposals, voted_on_more_than_5_pct_of_proposals, and voted_on_more_than_1_pct_of_proposals columns
organizations = organizations.drop(columns=['voted_on_more_than_10_pct_of_proposals', 'voted_on_more_than_5_pct_of_proposals', 'voted_on_more_than_1_pct_of_proposals'])
organizations
```

```python
# join proposals_per_organization
organizations = pd.merge(organizations, proposals_per_organization, left_index=True, right_index=True, how='left', validate='one_to_one')
organizations
```
