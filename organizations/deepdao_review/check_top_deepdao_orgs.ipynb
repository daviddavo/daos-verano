```python
import pandas as pd

df = pd.read_csv('top_100_deepdao_orgs.csv')
df.head()
```

```python
import duckdb
con = duckdb.connect(database=':memory:', read_only=False)
con.execute("CREATE VIEW deployments AS SELECT * FROM read_parquet('../parquets_version7/deployments.parquet')")
```

```python
def search_deployments(search: str, org_id='') -> pd.DataFrame:
    # search = "Gnosis"
    escaped_search = search.replace("'", "''")
    query = """
    SELECT
        *
    FROM
        deployments
    WHERE
        name ILIKE '%{}%'
        OR additional ILIKE '%{}%'
    """.format(escaped_search, escaped_search)
    results_df = con.execute(query).df()
    if org_id:
        # add deepdao_org_id as column
        results_df['organizationId'] = org_id
    return results_df
```

```python
search_results = []
# for row in df
for row in df.itertuples():
    name = row.daoName
    org_id = row.organizationId
    search_results.append(search_deployments(name, org_id))
    print(name, org_id)
```

```python
# concat the search_results list
results_df = pd.concat(search_results)
results_df.head()
```

```python
df.merge(results_df, on='organizationId', how='left').groupby('organizationId').id.count().sort_values(ascending=True).head(20)
```
