```python
# read votes.csv as df
import pandas as pd
df = pd.read_csv('votes.csv')
# drop unnamed col
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
# uintvote val counts
df.uintVote.value_counts()
```

```python
# member address val counts
df['memberAddress'].value_counts()
```

```python
# * platform: 'daohaus'
df['platform'] = 'daohaus'
# * rename molochAddress: platform_deployment_id
df = df.rename(columns={'molochAddress': 'platform_deployment_id'})
# * proposalAddress: proposal_id
df = df.rename(columns={'proposalAddress': 'proposal_id'})
# * id: vote_id
df = df.rename(columns={'id': 'vote_id'})
# * memberAddress: voter
df = df.rename(columns={'memberAddress': 'voter'})
# * createdAt: date
df = df.rename(columns={'createdAt': 'date'})
# * uintVote: choice
df = df.rename(columns={'uintVote': 'choice'})
# memberPower: weight
df = df.rename(columns={'memberPower': 'weight'})

df = df[['platform', 'platform_deployment_id', 'proposal_id', 'vote_id', 'voter', 'date', 'choice', 'weight']]
df.head()
```

```python
# make choice a string
df['choice'] = df['choice'].astype(str)
# make date a datetime, parse from seconds epoch
df['date'] = pd.to_datetime(df['date'], unit='s')
df.head()
```
