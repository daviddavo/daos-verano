```python
# open responses_by_chain_without_42161_or_10.pkl as responses_by_chain
import pickle
responses_by_chain = pickle.load(open('responses_by_chain_without_42161_or_10.pkl', 'rb'))
```

```python
# convert into a list of votes
all_votes = []
for chain_id in responses_by_chain:
    print(chain_id)
    for list_of_proposal_votes in responses_by_chain[chain_id]:
        for proposal in list_of_proposal_votes:
            votes = proposal['votes']
            if votes is not None:
                all_votes.extend(votes)
            
all_votes
```

```python
# load all_votes into a df
import pandas as pd
df = pd.DataFrame(all_votes)
df
```

```python
# make proposal id a column
df['proposal_id'] = df['proposal'].apply(lambda x: x['id'])
# make governor id a column
df['governor_id'] = df['proposal'].apply(lambda x: x['governor']['id'])
# count ids and keep the first governor id
votes_by_proposal = df.groupby('proposal_id').agg({'governor_id': 'first', 'id': 'count'})
votes_by_proposal
```

```python
# make a col chain_id that is the governor id before the second :
votes_by_proposal['chain_id'] = votes_by_proposal['governor_id'].apply(lambda x: x.split(':')[1])
votes_by_proposal
```

```python
# show only chain id 1
votes_by_proposal[votes_by_proposal['chain_id'] == '1'].sort_values('id', ascending=False)
```

```python
import json
chain_10_proposals = json.load(open('eip155:10/first_pass_proposals.json', 'r'))
# chain_10_proposals
# chain_10_proposals[1]
```

```python
# start with chain 10
# we want to read in the first_pass_proposals.json in the eip155:10 folder as a json
import json
chain_10_proposals = json.load(open('eip155:10/first_pass_proposals.json', 'r'))
first_pass_chain_10_votes = []
for dao in chain_10_proposals:
    for proposal in dao:
        first_pass_chain_10_votes.extend(proposal['votes'])
len(first_pass_chain_10_votes)
```

```python
first_pass_chain_10_votes_df = pd.DataFrame(first_pass_chain_10_votes)
first_pass_chain_10_votes_df['proposal_id'] = first_pass_chain_10_votes_df['proposal'].apply(lambda x: x['id'])
first_pass_chain_10_votes_df['governor_id'] = first_pass_chain_10_votes_df['proposal'].apply(lambda x: x['governor']['id'])
first_pass_chain_10_votes_df.head()
```

```python
# find all the json files in the eip155:10 folder
import glob
chain_10_json_filenames = glob.glob('eip155:10/*.json')
# remove the first_pass_proposals.json file
chain_10_json_filenames.remove('eip155:10/first_pass_proposals.json')
chain_10_json_filenames
```

```python
# the proposal id in the filename
proposal_ids_to_replace = [x.split('/')[1].split('.')[0] for x in chain_10_json_filenames]
proposal_ids_to_replace
```

```python
# how many votes are there that match the proposal ids to replace?
len(first_pass_chain_10_votes_df[first_pass_chain_10_votes_df['proposal_id'].isin(proposal_ids_to_replace)])
# this is correct; 3000 x 8 = 24000
```

```python
# remove the votes that match the proposal ids to replace
chain_10_votes_df = first_pass_chain_10_votes_df[~first_pass_chain_10_votes_df['proposal_id'].isin(proposal_ids_to_replace)]
```

```python
# read all of the json files in
chain_10_proposal_jsons_to_replace = []
for filename in chain_10_json_filenames:
    with open(filename, 'r') as f:
        chain_10_proposal_jsons_to_replace.append(json.load(f))
```

```python
big_chain_10_proposal_votes = []
for json_file in chain_10_proposal_jsons_to_replace:
    big_chain_10_proposal_votes.extend(json_file['votes'])
len(big_chain_10_proposal_votes) / 8 # avg votes per proposal
```

```python
# make new_chain_10_votes into a df
big_chain_10_proposal_votes_df = pd.DataFrame(big_chain_10_proposal_votes)
big_chain_10_proposal_votes_df['proposal_id'] = big_chain_10_proposal_votes_df['proposal'].apply(lambda x: x['id'])
big_chain_10_proposal_votes_df['governor_id'] = big_chain_10_proposal_votes_df['proposal'].apply(lambda x: x['governor']['id'])
big_chain_10_proposal_votes_df
```

```python
# * platform
df['platform'] = 'tally'
# * rename governor_id to platform_deployment_id
df.rename(columns={'governor_id': 'platform_deployment_id'}, inplace=True)
# * proposal_id no chg
# * id to vote_id
df.rename(columns={'id': 'vote_id'}, inplace=True)
# * voter -- the id of the person who made the vote
df['voter'] = df['voter'].apply(lambda x: x['id'])
# * block['timestamp'] to date
df['date'] = df['block'].apply(lambda x: x['timestamp'])
# * support to choice
df.rename(columns={'support': 'choice'}, inplace=True)
# * weight -- the vote weight, no chg

# keep only these cols
df = df[['platform', 'platform_deployment_id', 'proposal_id', 'vote_id', 'voter', 'date', 'choice', 'weight']]
df.head()
```

```python
# make date a pandas date, parse from 2022-12-16T12:27:00Z	
df['date'] = pd.to_datetime(df['date'])
# remove tz
df['date'] = df['date'].dt.tz_localize(None)
df.head()
```
