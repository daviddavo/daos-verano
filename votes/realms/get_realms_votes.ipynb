```python
# glob get all of the json files in ../realms_node/output_votes_2023_6_12
# read them into dfs and concat them into one df

import glob
import pandas as pd

files = glob.glob('../../aug3/output_votes_2023_8_3/*.json')
len(files)
```

```python
df = pd.concat([pd.read_json(f) for f in files], ignore_index=True)
df.shape
```

```python
df.head()
```

```python
# expand account column
df = pd.concat([df.drop(['account'], axis=1), df['account'].apply(pd.Series)], axis=1)
df.head()
```

```python
df.shape
```

```python
# keep only where isRelinquished is False
df = df[df['isRelinquished'] == False]
df.shape
```

```python
# count number of unique values in each column except 'vote' and 'voteWeight'
df[['pubkey', 'owner', 'accountType', 'proposal', 'governingTokenOwner',
       'isRelinquished', 'voterWeight']].nunique()
```

```python
# expand the 'vote' column
df = pd.concat([df.drop(['vote'], axis=1), df['vote'].apply(pd.Series)], axis=1)
df.head()
```

```python
df['voterWeight'].describe()
```

```python
# make voter weight into a float
df['voterWeight'] = df['voterWeight'].astype(float)
```

```python
# voteWeight describe
df['voteWeight'].describe()
```

```python
# * platform
df['platform'] = 'realms'
# * platform_deployment_id TKTKTK, need to merge w the proposals data
# * proposal rename to proposal_id
# df.rename(columns={'proposal': 'proposal_id'}, inplace=True)
df['proposal_id'] = df['owner'] + '_' + df['proposal']
# * pubkey to vote_id
df.rename(columns={'pubkey': 'vote_id'}, inplace=True)
# * governingTokenOwner to voter
df.rename(columns={'governingTokenOwner': 'voter'}, inplace=True)
# * date doesn't exist, fill nans
df['date'] = pd.NaT
# * choice
df.rename(columns={'approveChoices': 'choice'}, inplace=True)
# * voterWeight to weight
df.rename(columns={'voterWeight': 'weight'}, inplace=True)
df['vote_id'] = df['owner'] + '_' + df['vote_id']

df = df[['platform', 'proposal_id', 'vote_id', 'voter', 'date', 'choice', 'weight']]
df.head()
```

```python
proposals_df = pd.read_csv('../../proposals/realms/realms_deployment_proposal_ids.csv')
proposals_df.head()
```

```python
proposals_df = proposals_df[['platform_deployment_id', 'proposal_id']]
proposals_df.head()
```

```python
# merge df <- proposals_df
df = pd.merge(df, proposals_df, on='proposal_id', how='left', validate='m:1')
df.head()
```

```python
# how many null values in platform_deployment_id
df['platform_deployment_id'].isnull().sum()
```

```python
# reorder cols
df = df[['platform', 'platform_deployment_id', 'proposal_id', 'vote_id', 'voter', 'date', 'choice', 'weight']]
df.head()
```

```python
df.shape
```

```python
df['id_to_hash'] = df['platform'] + '_' + df['platform_deployment_id'] + '_' + df['proposal_id'] + '_' + df['vote_id']
df.head()
```

```python
import uuid
df['id'] = df['id_to_hash'].apply(lambda x: str(uuid.uuid5(uuid.NAMESPACE_OID, str(x))))
df.head()
```

```python
df['proposal_id_to_hash'] = df['platform'] + '_' + df['platform_deployment_id'] + '_' + df['proposal_id']
df['proposal_id'] = df['proposal_id_to_hash'].apply(lambda x: str(uuid.uuid5(uuid.NAMESPACE_OID, str(x))))
df.head()
```

```python
# rename vote_id to platform_vote_id
df.rename(columns={'vote_id': 'platform_vote_id'}, inplace=True)
df.columns
```

```python
df['deployment_id_to_hash'] = df['platform'] + '_' + df['platform_deployment_id']
df['deployment_id'] = df['deployment_id_to_hash'].apply(lambda x: str(uuid.uuid5(uuid.NAMESPACE_OID, str(x))))
df.head()
```

```python
df.columns
```

```python
df = df[[
    'id',
    'proposal_id',
    'deployment_id',
    'platform_vote_id',
    'voter',
    'date',
    'choice',
    'weight'
]]
df.head()
```

```python
# save to csv
df.to_csv('realms_votes.csv', index=False)
```
