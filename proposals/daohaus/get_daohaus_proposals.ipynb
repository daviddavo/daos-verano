```python
import pandas as pd

# read proposals.csv
df = pd.read_csv('proposals.csv')
# drop unnamed column
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
import json

def tryParseDetails(details):
    title, desc = "", ""
    
    try:
        j = json.loads(details)

        if isinstance(j, dict):
            title = j.get('title')
            desc = j.get('description')
        else:
            desc = j
    except json.JSONDecodeError:
        desc = details

    return pd.Series({'title': title, 'description': desc})
    
df[['title', 'description']] = df['details'].astype(str).apply(tryParseDetails)
```

```python
# make a votes_count column from yes_votes and no_votes sum
df['votes_count'] = df['yesVotes'] + df['noVotes']
```

```python
# * platform
df['platform'] = 'daohaus'
# * platform_deployment_id
# rename molochAddress to platform_deployment_id
df = df.rename(columns={'molochAddress': 'platform_deployment_id'})
# * proposal_id
# rename id to proposal_id
df = df.rename(columns={'id': 'proposal_id'})
# * author
# rename proposer to author
df = df.rename(columns={'proposer': 'author'})

# * date -- this should be as a pandas date, not as an epoch
# convert epoch to pandas date
df['date'] = pd.to_datetime(df['createdAt'], unit='s')
df['start'] = df['date']
df['end'] = pd.to_datetime(df['processedAt'], unit='s')
# * votes_count
```

```python
df = df[['platform', 'platform_deployment_id', 'proposal_id', 'author', 'date', 'votes_count', 'title', 'description', 'start', 'end']]
df.head()
```

```python
# save as csv
# df.to_csv('daohaus_proposals.csv', index=False)
df.to_parquet('daohaus_proposals.parquet', index=False)
```
