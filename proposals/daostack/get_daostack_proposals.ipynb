```python
import pandas as pd

df = pd.read_csv('proposals.csv')
# drop unknown col
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df.iloc[2]
```

```python
# * platform
df['platform'] = 'daostack'
# * platform_deployment_id
# rename dao to platform_deployment_id
df = df.rename(columns={'dao': 'platform_deployment_id'})
# * proposal_id
# id -> proposal_id
df = df.rename(columns={'id': 'proposal_id'})
# * author
# proposer -> author
df = df.rename(columns={'proposer': 'author'})
# * date -- this should be as a pandas date, not as an epoch
# createdAt -> date
df['date'] = pd.to_datetime(df['createdAt'], unit='s')
# createdAt -> start
df['start'] = df['date']
# executedAt -> end
df['end'] = pd.to_datetime(df['executedAt'], unit='s')
# * votes_count
# tbd
df.head()
```

```python
df = df[['platform', 'platform_deployment_id', 'proposal_id', 'author', 'date', 'title', 'description', 'start', 'end']]
df.head()
```

```python
# add in votes
votes_df = pd.read_csv('votes.csv')
votes_df.head()
```

```python
votes_by_proposal = votes_df.groupby('proposal').id.count()
# reindex
votes_by_proposal = votes_by_proposal.reset_index()
# name the col votes_count
votes_by_proposal = votes_by_proposal.rename(columns={'proposal': 'proposal_id', 'id': 'votes_count'})
votes_by_proposal.head()
```

```python
# left join
df = df.merge(votes_by_proposal, on='proposal_id', how='left', validate='one_to_one')
df.head()
```

```python
# fill na with 0
df['votes_count'] = df['votes_count'].fillna(0)
# make votes_count an int
df['votes_count'] = df['votes_count'].astype(int)
df.head()
```

```python
df = df[['platform', 'platform_deployment_id', 'proposal_id', 'author', 'date', 'votes_count', 'title', 'description', 'start', 'end']]
```

```python
# save
# df.to_csv('daostack_proposals.csv', index=False)
df.to_parquet('daostack_proposals.parquet', index=False)
```
