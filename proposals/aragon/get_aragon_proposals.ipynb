```python
import pandas as pd
# Note that aragon votes are actually proposals; votes are called "casts"
proposals_df = pd.read_csv('votes.csv')
# remove unnamed column
proposals_df = proposals_df.loc[:, ~proposals_df.columns.str.contains('^Unnamed')]
proposals_df.head()
```

```python
proposals_df.info()
```

```python
# rename orgAddress to platform_id
proposals_df = proposals_df.rename(columns={'orgAddress': 'platform_deployment_id'})
# add column platform with value 'aragon'
proposals_df['platform'] = 'aragon'
# rename id to proposal_id
proposals_df = proposals_df.rename(columns={'id': 'proposal_id'})
# rename creator to author
proposals_df = proposals_df.rename(columns={'originalCreator': 'author'})
# rename startDate to created_at
proposals_df = proposals_df.rename(columns={'startDate': 'date'})
# rename voteNum to votes_count
proposals_df = proposals_df.rename(columns={'voteNum': 'votes_count'})
# adding proposals start
proposals_df['start'] = proposals_df['date']
# rename executedAt to end
proposals_df = proposals_df.rename(columns={'executedAt': 'end'})
# adding proposals title
proposals_df['title'] = pd.NA
proposals_df['title'] = proposals_df['title'].astype(str)
# rename metadata to description
proposals_df = proposals_df.rename(columns={'metadata': 'description'})
proposals_df.head()
```

```python
proposals_df = proposals_df[['platform', 'platform_deployment_id', 'proposal_id', 'author', 'date', 'votes_count', 'title', 'description', 'start', 'end']]
proposals_df.head()
```

```python
# parse date from epoch
for c in ['date', 'start', 'end']:
    proposals_df[c] = pd.to_datetime(proposals_df[c], unit='s')
proposals_df.head()
```
