# Snapshot

```python
snapshot_proposals_df = pd.read_csv('02_snapshot_proposals/1687429408_snapshot_proposals.csv')
snapshot_proposals_df
```

```python
_pre_merge_deepdao_daos_df = deepdao_daos_df.drop(columns=['platformId', 'website', 'mostRecentProposalDate', 'aum'])
```

```python
snapshot_proposals_df.columns
```

```python
_pre_merge_snapshot_proposals_df = snapshot_proposals_df[[
    'deepdaoDaoId',
    'snapshotSpaceId',
    'snapshotProposalId',
    'author',
    'created',
    'votes',
]]
# rename snapshotSpaceId to platformSpaceId
_pre_merge_snapshot_proposals_df = _pre_merge_snapshot_proposals_df.rename(columns={'snapshotSpaceId': 'platformSpaceId'})
# rename snapshotProposalId to platformProposalId
_pre_merge_snapshot_proposals_df = _pre_merge_snapshot_proposals_df.rename(columns={'snapshotProposalId': 'platformProposalId'})
_pre_merge_snapshot_proposals_df
```

```python
deepdao_daos_df_with_snapshot_proposals = pd.merge(
    _pre_merge_deepdao_daos_df,
    _pre_merge_snapshot_proposals_df,
    how='right',
    on='deepdaoDaoId',
    validate='one_to_many'
)
deepdao_daos_df_with_snapshot_proposals
```

```python
deepdao_daos_df_with_snapshot_proposals.groupby(['deepdaoOrganizationId', 'platformTitle', 'platformSpaceId']).size()
```

```python
# find the rows where platformTitle is Aragon
deepdao_daos_df[deepdao_daos_df.platformTitle == 'Aragon']
```

```python
# let's try to match these to the Kaggle DAO Analyzer organizations.csv data
aragon_organizations_df = pd.read_csv('daoanalyzer/organizations.csv')
# drop unnamed column
aragon_organizations_df = aragon_organizations_df.drop(columns=['Unnamed: 0'])
aragon_organizations_df
```

```python
# inner join on the name column for shits and giggles
pd.merge(
    deepdao_daos_df[deepdao_daos_df.platformTitle == 'Aragon'],
    aragon_organizations_df,
    how='inner',
    left_on='name',
    right_on='name',
    # validate='one_to_one'
).shape
```

```python
# get unique aragon organization names
aragon_organizations_df.name.unique()
```

```python
# get the deepdao daos that are aragon organizations
deepdao_daos_df[deepdao_daos_df.platformTitle == 'Aragon'].name
```

```python
# maybe we should use all of the aragon DAOs and backfill the deepdao data
# in the cases where they are not affiliated with a deepdao organization, we assume they are
# in their own organization
```

```python
# let's work with the direct matches for now
names = ['Airalab', 'Aavegotchi', 'PieDAO', 'NFTX']
aragon_orgs_to_merge = aragon_organizations_df.loc[aragon_organizations_df.name.isin(names)]
aragon_orgs_to_merge
```

```python
aragon_deepdao_daos_df = deepdao_daos_df[deepdao_daos_df.platformTitle == 'Aragon']
aragon_deepdao_daos_df
```

```python
# rename Aavegotchi DAO in aragon_deepdao_daos_df to Aavegotchi
aragon_deepdao_daos_df.loc[aragon_deepdao_daos_df.name == 'Aavegotchi DAO', 'name'] = 'Aavegotchi'

# merge with deepdao daos
deepdao_orgs_with_aragon_addresses_df = pd.merge(
    aragon_orgs_to_merge[['name', 'orgAddress']],
    aragon_deepdao_daos_df[['deepdaoDaoId', 'deepdaoOrganizationId', 'name']],
    on='name',
    how='left',
    validate='one_to_one'
)
deepdao_orgs_with_aragon_addresses_df
```

## Getting Aragon proposals

```python
# open the aragon votes csv
aragon_proposals_df = pd.read_csv('daoanalyzer/votes.csv')
# drop unnamed column
aragon_proposals_df = aragon_proposals_df.drop(columns=['Unnamed: 0'])
aragon_proposals_df
```

```python
# merge the aragon proposals with the aragon organizations inner join
aragon_proposals_df_with_orgs = pd.merge(
    aragon_proposals_df,
    aragon_orgs_to_merge,
    how='inner',
    left_on='orgAddress',
    right_on='orgAddress',
    validate='many_to_one'
)
aragon_proposals_df_with_orgs
```

```python
aragon_proposals_df_with_orgs.columns
```

```python
aragon_proposals_df_with_orgs = aragon_proposals_df_with_orgs[[
    'id_x',
    'orgAddress',
    'creator',
    'createdAt',
    'startDate',
    'appAddress',
    'voteNum'
]]
# make id_x the id
aragon_proposals_df_with_orgs = aragon_proposals_df_with_orgs.rename(columns={'id_x': 'id'})
aragon_proposals_df_with_orgs
```

```python
aragon_proposals_df_with_orgs = pd.merge(
    aragon_proposals_df_with_orgs,
    deepdao_orgs_with_aragon_addresses_df,
    how='left',
    on='orgAddress',
    validate='many_to_one'
)
aragon_proposals_df_with_orgs
```

```python
_pre_merge_aragon_proposals_df_with_orgs = aragon_proposals_df_with_orgs[[
    'deepdaoDaoId',
    'deepdaoOrganizationId',
    'orgAddress', # platformSpaceId
    'appAddress', # platformProposalId
    'creator', # author
    'voteNum', # votes
    'createdAt', # created
]]
_pre_merge_aragon_proposals_df_with_orgs = _pre_merge_aragon_proposals_df_with_orgs.rename(columns={'orgAddress': 'platformSpaceId'})
_pre_merge_aragon_proposals_df_with_orgs = _pre_merge_aragon_proposals_df_with_orgs.rename(columns={'voteNum': 'votes'})
_pre_merge_aragon_proposals_df_with_orgs = _pre_merge_aragon_proposals_df_with_orgs.rename(columns={'appAddress': 'platformProposalId'})
_pre_merge_aragon_proposals_df_with_orgs = _pre_merge_aragon_proposals_df_with_orgs.rename(columns={'creator': 'author'})
_pre_merge_aragon_proposals_df_with_orgs = _pre_merge_aragon_proposals_df_with_orgs.rename(columns={'createdAt': 'created'})
# make created a datetime
_pre_merge_aragon_proposals_df_with_orgs['created'] = pd.to_datetime(_pre_merge_aragon_proposals_df_with_orgs['created'], unit='s')

# make platformTitle Aragon
_pre_merge_aragon_proposals_df_with_orgs['platformTitle'] = 'Aragon'
_pre_merge_aragon_proposals_df_with_orgs
    
```

```python
deepdao_daos_df_with_proposals
```

```python
# combine the aragon proposals with the snapshot proposals
deepdao_daos_df_with_proposals = pd.concat([
    deepdao_daos_df_with_proposals,
    _pre_merge_aragon_proposals_df_with_orgs
])
deepdao_daos_df_with_proposals
```
